```python
import pandas as pd
```

```python
df = pd.read_csv("hydroelectric_power_station.csv")
```

```python


for i in df['water_volume_billion.m3']:
    
    df['water_volume_billion.m3'] = df['water_volume_billion.m3'].replace(i, i.replace(',', ''))
```

```python
new_df = df
```

```python
new_df = df.dropna()
```

```python
new_df.tail()
```

```python
new_df['water_volume_billion.m3'].dtype
```

```python
X = new_df[['date', 'average_daily_flow_m3/s', 'average_daily_consumption_m3/s', 'cloud', 'rainfall', 'tem_day', 'tem_night']]
```

```python
y = new_df['water_volume_billion.m3']
```

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
#plt.plot(X['date'], y)
fig = plt.figure(figsize = (15, 15))
plt.plot(X['date'], X['rainfall'])
plt.plot(X['date'], X['average_daily_flow_m3/s'])
plt.plot(X['date'], X['average_daily_consumption_m3/s'])
plt.plot(X['date'], X['cloud'])
plt.plot(X['date'], X['tem_day'])
plt.plot(X['date'], X['tem_night'])

plt.show()
```

```python
plt.plot(X['date'], y)
plt.show()
```

```python
myModel2 = LinearRegression()
```

```python
myModel2.fit(X, y)
```
