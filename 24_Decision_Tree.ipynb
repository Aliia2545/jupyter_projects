```python
from sklearn.linear_model import LogisticRegression
```

```python
iris = pd.read_csv('iris (2).csv', sep = ',')
iris.head()
```

```python
X = iris[['PetalLengthCm', 'PetalWidthCm']]
y = iris['Species']
```

```python
model = DecisionTreeClassifier(criterion= 'entropy', max_depth = 2)

model.fit(X, y)
```

```python
model2 = LogisticRegression()
model2.fit(X, y)
```

```python
iris['Species'] = iris['Species'].map({'Iris-setosa': 0,
                                      'Iris-versicolor': 1,
                                      'Iris-virginica': 2})
```

На графике каждая точка это один цветок и по оси Х - длина лепестка, по оси Y - ширина лепестка.

Как через условия (if-else) сделать правило для разделения цветков?

Попробуем построить дерево:

```python
test_list = []
train_list = []
for _ in range(25):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
    model.fit(X_train, y_train)
#     model.score(X_train, y_train)
#     model.score(X_test, y_test)
    test_list.append(round(model.score(X_test, y_test), 2))
    train_list.append(round(model.score(X_train, y_train), 2))
print(test_list, "test list")
print(train_list, "train list")

```

```python
a = np.array(test_list)
print(a.mean(), "test mean")
b = np.array(train_list)
print(b.mean(), "train mean")
```
